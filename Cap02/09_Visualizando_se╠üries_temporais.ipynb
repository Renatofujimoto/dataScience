```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns


%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

### Gráfico de série temporal

```python
caminho = "C:/Users/renat/OneDrive/Área de Trabalho/DataScience/dados/Superstore-Sales.csv"

df = pd.read_csv(caminho, index_col="Order Date", parse_dates=True, encoding="ISO-8859-1")
df.head(3)
```

```python
df["Order Quantity"].plot()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0) #n = é a quantidade de observações

plt.xlabel("Order Date")
plt.ylabel("Order Quantity")
plt.title("Vendas")

df2["Order Quantity"].plot()
```
